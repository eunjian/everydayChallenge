# Sum Calculator

Typical calculators can display only one calculation result at a time. Consequently, when multiple calculation results are needed, it's cumbersome to keep track of previous results separately.
To address this issue, the player has developed a sum calculator. The sum calculator has the capability to receive multiple calculations($T$) as input and then output the sum of all individual calculation results. 

The input expressions for the sum calculator must satisfy the following conditions:
- The calculation expression follows the format: `<integer> <operator> <integer>`.
- \<operator> can be one of the four arithmetic operators: addition, subtraction, multiplication, or division. The remainder from division is discarded.

Given a set number of calculation($T$) expressions for the sum calculator, determine the final output result.

**Input:**

- The first line contains the number of expressions, $T$, and $1 \le T \le 100$.
- The next lines each contain a calculation expression.
- Each calculation expression follows the format: `<integer> <operator> <integer>`.
- \<integer> will be an integer ranging from 1 to 1000. 
- \<operator> will be one of the characters: +, -, *, /. Each representing addition, subtraction, multiplication, or division respectively.

**Output:**

Print the result of calculations obtained from the sum calculator.

## Example1:

Input:
```
3
1 + 3
4 / 3
3 - 2
```
    
Output
```
6
```
    
## Example 2
    
Input:
```
3
3 - 4
4 / 5
5 * 1
```
    
Output:
```
4
```
    
â‹‡ Please keep the input and output formats carefully

````python
t = int(input())  # Number of calculations

# Each formula has the format: <int> <operator> <int>
c = [input() for _ in range(t)]  # Input each formula

result = []  # Store results of each formula

for f in c:
    i1, operator, i2 = f.split()
    
    if operator == '+':
        s = int(i1) + int(i2)
        result.append(s)  # Store the current result
    elif operator == '-':
        s = int(i1) - int(i2)
        result.append(s)
    elif operator == '*':
        s = int(i1) * int(i2)
        result.append(s)
    elif operator == '/':
        s = int(i1) // int(i2)  # Discard the remainder of division
        result.append(s)
    else:
        print("Wrong input")

# Sum up all the results and print the total
print(sum(result))
````
